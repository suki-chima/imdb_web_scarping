## Web Scraping example

```python
#Import libraries
from bs4 import BeautifulSoup
import requests
import csv
```

```python
#Open a new file
file = open('imdb_webscrapes.csv', 'w')
#We are going to write into our CSV file
writer = csv.writer(file)
```

```python
#Create a header for the row of the csv file
writer.writerow(['film', 'rating'])
```

```python
#Request webpage and store it against a variable
page_to_scrape = requests.get("https://www.imdb.com/search/title/?sort=num_votes,desc&start=1&title_type=feature&year=1950,2012")
```

```python
page_to_scrape
```

```python
#Use BeaustifulSoup to parse the HTML code and store against a variable
soup = BeautifulSoup(page_to_scrape.text, 'html.parser')
```

```python
soup
```

```python
"""
Find all the items in the page with the rating
"""
ratings= soup.findAll('strong')
```

```python
ratings
```

```python
"""
Find all the items in the page with the title of the movie
 """
titles = soup.findAll('h3', attrs={"class":"lister-item-header"})
```

```python
titles
```

```python
for film in titles:
    print(film.text)
```

```python
#Loop through both lists, print and format the results
for rating in ratings:
    print(rating.text)
for title in titles:
    print(title.text)
```

```python
#Loop through both lists using Zip function to combime them together
for rating, title in zip(ratings, titles):
    print(title.text + "-" + rating.text)
    #Write each item onto a new row
    writer.writerow([title.text, rating.text])
```

```python
#Close the csv file
file.close()
```
